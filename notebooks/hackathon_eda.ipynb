## Importing geospatial data

```python
import geopandas as gpd
```

```python
usa_shape = gpd.read_file(pathlib.Path(DATASETS, 'usa.shp'))
```

```python
usa_shape.head()
```

```python
# Need to project geometries to a CRS - EPSG = 3035
usa_proj = usa_shape.to_crs(epsg=3035)
```

```python
# Find areas by geoid10, SF and CF
geo_area = usa_proj.dissolve(by='GEOID10')['geometry'].area
sf_area = usa_proj.dissolve(by='SF')['geometry'].area
cf_area = usa_proj.dissolve(by='CF')['geometry'].area
```

```python
geo_area.name = 'geo_area'
sf_area.name = 'sf_area'
cf_area.name = 'cf_area'
```

```python
usa_proj = pd.merge(usa_proj, geo_area, on='GEOID10')
usa_proj = pd.merge(usa_proj, sf_area, on='SF')
usa_proj = pd.merge(usa_proj, cf_area, on='CF')
```

```python
usa_proj.drop(columns=['geo_size'], inplace=True)
```

```python
usa_proj.to_file('datasets/usa_projected_3035.shp')
```

```python
## Projection to m^2
usa_proj_m = usa_shape.to_crs(epsg=6933)
```

```python
geo_area_m = usa_proj_m.dissolve(by='GEOID10')['geometry'].area
sf_area_m = usa_proj_m.dissolve(by='SF')['geometry'].area
cf_area_m = usa_proj_m.dissolve(by='CF')['geometry'].area

geo_area_m.name = 'geo_area'
sf_area_m.name = 'sf_area'
cf_area_m.name = 'cf_area'

usa_proj_m = pd.merge(usa_proj_m, geo_area_m, on='GEOID10')
usa_proj_m = pd.merge(usa_proj_m, sf_area_m, on='SF')
usa_proj_m = pd.merge(usa_proj_m, cf_area_m, on='CF')
```

```python
usa_proj_m.to_file('datasets/usa_projected_6933.shp')
```

```python
usa_proj_m.head()
```

```python
## Joining shape areas onto original dataset

# Set index of each dataset
communities.set_index(['Census tract ID', 'County Name', 'State/Territory'], inplace=True)
usa_proj_m.rename(columns={'GEOID10':'Census tract ID', 'CF': 'County Name', 'SF': 'State/Territory'}, inplace=True)
usa_proj_m.set_index(['Census tract ID', 'County Name', 'State/Territory'], inplace=True)
```

```python
usa_proj_m.reset_index(inplace=True)

usa_proj_m['Census tract ID'] = usa_proj_m['Census tract ID'].astype(int)
usa_proj_m.set_index(['Census tract ID', 'County Name', 'State/Territory'], inplace=True)
```

```python
# Join dataset to each area of state, county and census tract
communities = communities.join(usa_proj_m.loc[:,['geo_area', 'sf_area', 'cf_area']])
```

```python
# Convert area to square km instead of square m
communities.loc[:,['geo_area', 'sf_area', 'cf_area']] = communities.loc[:,['geo_area', 'sf_area', 'cf_area']] / 1e6
```

```python
communities.to_csv('communities_inc_area.csv')
```

## Exporting as csv file to reduce size - using point centers

```python
usa_shape['rep_point'] = usa_shape['geometry'].representative_point()
usa_shape['x_coord'] = usa_shape['rep_point'].x
usa_shape['y_coord'] = usa_shape['rep_point'].y
usa_shape.rename(columns={'y_coord':'lat', 'x_coord': 'lon'}, inplace=True)
```

```python
### IMPORTANT - to_json using orient='records
usa_shape[['GEOID10', 'CF', 'SF', 'lat', 'lon', 'geo_size']].to_json('centres_json.json', orient='records')
usa_shape[['GEOID10', 'CF', 'SF', 'lat', 'lon', 'geo_size']].to_csv('census_tract_centres.csv')
```

```python
counties = usa_shape.dissolve(by='CF')
```

```python
counties.drop(columns=['rep_point', 'lon', 'lat'], inplace=True)
```

```python
counties['county_rep_point'] = counties['geometry'].representative_point()
counties['x_coord'] = counties['county_rep_point'].x
counties['y_coord'] = counties['county_rep_point'].y
counties.rename(columns={'y_coord':'lat', 'x_coord': 'lon'}, inplace=True)
```

```python
counties[['SF', 'lat', 'lon', 'geo_size']].reset_index().to_csv('county_centres.csv')
```

```python
states = usa_shape.dissolve(by='SF')
states.drop(columns=['rep_point', 'lon', 'lat'], inplace=True)
states['county_rep_point'] = states['geometry'].representative_point()
states['x_coord'] = states['county_rep_point'].x
states['y_coord'] = states['county_rep_point'].y
states.rename(columns={'y_coord':'lat', 'x_coord': 'lon'}, inplace=True)
```
